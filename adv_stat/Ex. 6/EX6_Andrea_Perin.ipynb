### EX 1

A particle detector has a $\mu$ identification efficiency of 95%, and a probability of identifying a $\pi$ as
a $\mu$ of 2%. If a particle is identified as a $\mu$, a trigger is fired. Knowing that the particle beam is a
mixture of 90% $\pi$s and 10% $\mu$s, what is the probability that a trigger is really fired by a $\mu$ ? What
is the signal-to-noise (S/N) ratio ?

````r
eff <- 0.95
false_positive <- 0.02
pi_comp <- 0.9
mu_comp <- 0.1

real_trigger <- (mu_comp*eff)/(mu_comp*eff+pi_comp*false_positive)

SNR <-real_trigger/(1-real_trigger)

cat("The probability that a trigger is fired by a mu is",real_trigger)
cat("\nThe SNR is",SNR)
````

### EX 4

Three boxes contain two rings each, but in one of them they are both gold, in the second both silver, and in the third one of each type. You have the choice of randomly extracting a ring from one of the boxes, the content of which is unknown to you. You look at the selected ring, and you then have the possibility of extracting a second ring, again from any of the three boxes. Let us assume the first ring you extract is gold. Is it then preferable to extract the second one from the same or from a different box?

Events at hand:
* getting gold (G) drawing the second ring from the same box;
* getting G drawing the second ring from a different box.

Named $B_0, B_1, B_2$ the boxes with respectively 0, 1 and 2 gold rings, then:
* Probability of having drawn $G$ from $B_0$: $P(B_0|G) = 0$;

* Probability of having drawn $G$ from $B_1$: $P(B_1|G) = \frac{1}{3}$;

* Probability of having drawn $G$ from $B_2$: $P(B_2|G) = \frac{2}{3}$.

As such, the probability of getting a second gold by changing box is:
$$
P(GG| \text{change of box}) = P(B_1|G) \cdot P(G|B_2 \text{ or } B_0) + P(B_2|G) \cdot P(G|B_1 \text{ or } B_0) = \frac{1}{3}\cdot \frac{1}{2} + \frac{2}{3} \cdot\frac{1}{4} = \frac{1}{3}
$$
while the probability of getting a second gold by not changing box is:
$$
P(GG| \text{ no change }) = P(B_1|G) \cdot P(\text{ a second }G|B_1) + P(B_2|G) \cdot P(\text{ a second }G|B_2) = \frac{1}{3}\cdot 0 + \frac{2}{3}\cdot 1 = \frac{2}{3}
$$
It is then preferable to not change box.

### EX 5

Estimate the following three dimensional integral
$$
\int_0^2 \int_3^6 \int_{-1}^1 \left(yx^2 + z\log y + e^x\right) dx \hspace{0.5em} dy\hspace{0.5em} dz
$$
using at least two Monte Carlo methods.

````r
#USING HIT OR MISS

func <- function(x,y,z) {
    return (y*x**2+z*log(y)+exp(x))
}

hit_or_miss <- function(n) {

    M <- func(2,6,1) #maximum value
    m <- func(0,6,-1) #minimum value
    
    #the boundaries
    x_low <- 0
    x_high <- 2
    y_low <- 3
    y_high <- 6
    z_low <- -1
    z_high <- 1
    
    #generating numbers in the domani (which is  a rectangle)
    x_vals <- runif(n, min=x_low, max=x_high)
    y_vals <- runif(n, min=y_low, max=y_high)
    z_vals <- runif(n, min=z_low, max=z_high)
        
    # raising the function by m, so that it entirely lies in the positives; 
    # the result will then be shifted downwards by the
    # appropriate quantity
    f_vals <- func(x_vals, y_vals, z_vals)-m
    
    #generatin random numbers in the (shifted) interval
    comparison <- runif(n, min=0, max=M-m)
        
    # counting the number of points lying below the function 
    #(no need of further checks, the function has been shifted)
    val <- sum(ifelse(comparison<=f_vals,
                     1,
                     0))
 
    #the rectangle containing the function
    const <- (M-m)*(x_high-x_low)*(y_high-y_low)*(z_high-z_low) 

    # usual hit or miss stuff
    val <- (val*const)/n
    
    #subtracting the value corresponding to the upward shifting
    neg_part <- m*(x_high-x_low)*(y_high-y_low)*(z_high-z_low)

    #the result is the following
    return(val+neg_part)
}

cat("The value of the integral is: ", hit_or_miss(1e3))
````

````r
# uncomment the following line if "faraway" is not yet installed
# install.packages("MASS", repos='http://cran.us.r-project.org')
library('MASS')
````

````r
# USING MCMC AND IMPORTANCE SAMPLING
sample_func <- function(x){
    ifelse(x[1]>=0 & x[1]<=2 & x[2]>=3 & x[2]<=6 & x[3]>=-1 & x[3]<=1,
                return((x[2]*x[1]*x[1]+x[3]*(log(3)+x[2]/3-1)+1+x[1]+x[1]**2/2)/104),
                return(0))
}

func <- function(x) {
    return(x[2]*x[1]**2+x[3]*log(x[2])+exp(x[1]))
}

# DRAWING POIINTS THROUGH METROPOLIS
metropolis <- function(func, theta_init, sigma, num) {
    num_feat <- length(theta_init)
    theta_cur <- theta_init
    func_Cur <- func(theta_cur)    
    func_Samp <- matrix(data=NA, nrow=num, ncol=num_feat+2)
    n_accept <- 0

    for (n in 1:num) {
        theta_prop <- mvrnorm(1, theta_cur, sigma) #draw from multivariate normal
        func_Prop <- func(theta_prop)
        logMR <- func_Prop - func_Cur  # Log10 of the Metropolis ratio
        if( logMR>=0 || logMR>log10(runif(1)) ) {
            theta_cur <- theta_prop
            func_Cur <- func_Prop
            n_accept <- n_accept + 1
            func_Samp[n, num_feat+2] <- 1 #denotes when a point is new
        }
        func_Samp[n, 1] <- func_Cur
        func_Samp[n, 2:(num_feat+1)] <- theta_cur
    }
    return(func_Samp)
}

# THE REAL IMPORTANCE SAMPLING
ratio <- function(points, f){
    l1 <- nrow(points)
    l2 <- ncol(points)
    num <- sapply(1:l1, function(i) f(points[i,2:l2])/points[i,1])
    return(mean(num))
}
````

````r
# initialization of the "parameters" of the chain
init <- c(runif(1, min=0, max=2),runif(1, min=3, max=6),runif(1, min=-1, max=1))
# selecting the sigma
sigma <- diag(0.001,3)
# choosing how many steps to take
num <- 1e4

# creating the points
sim <- metropolis(sample_func, init, sigma, num)
#only keeping relevant points
#pts <- sim[sim[,5]==1 & sim[,1]!=0, 1:4]
pts <- sim[sim[,1]!=0, 1:4]


cat("The value of the integral is ", ratio(pts, func))
````

### EX 6

Estimate the area under the curve $y = e^{(x+1)^2}$ and inside a triangle with vertices $V_1(1, 0),  V_2(0, 1)$
and $V_3(1, 0)$ with a Monte Carlo algorithm.

````r
func <- function(x) {
    ifelse(x>=-1 && x<=1,
        val <- min(1-abs(x), exp(-(x+1)**2)),
        val <- NaN)
    return(val)
}
func <- Vectorize(func)

area <- function(n) {
    x_min <- -1
    x_max <- 1
    y_max <- 0.655 #slightly above the crossing point of the triangle and the gaussian

    x_vals <- runif(n, min=x_min, max=x_max)
    f_vals <- func(x_vals)
    y_vals <- runif(n, min=0, max=y_max)

    area <- sum(y_vals<f_vals)*(x_max-x_min)*(y_max)/n
    return(area)
}

cat("The area under the curve is",area(1e5))
````

The value of $\approx 0.52$ is consistent with a rough eye estimate: by superimposing a triangle with vertices at $P_1(-1,0), P_2(1,0), P_3(0.345,0.655)$, we get an area of approximately $2\cdot0.5\cdot0.6 \approx 0.6$.

### EX 7

Given a region in the $x  y$ plane defined by the inequalities
$$
\Omega =
\begin{cases}
    \frac{1}{3} \leq 3x \leq 9-y \\
    \sqrt{x} \leq y \leq 3
\end{cases}
$$
estimate the integral
$$
\int_\Omega \left( e^x + \cos xy\right) dx \hspace{0.5em} dy
$$

````r
# some function defining the domain
domain_up <- function(x) {
    ifelse(1/9 <= x & x<=55/18-sqrt(109)/18,
        val <- sqrt(x),
        val <- NA)
    return(val)
}
domain_up <- Vectorize(domain_up)

domain_down <- function(x) {
    ifelse(1/9 <= x & x<=55/18-sqrt(109)/18,
        val <- min(9-3*x, 3),
        val <- NA)
    return(val)
}
domain_down <- Vectorize(domain_down)

# creating the appropriate matrix, containing the values of the function for each point
func <- function(x,y) {
    ifelse(1/9<=x & x<=55/18-sqrt(109)/18 & y<=3 & y<=9-3*x & y>= sqrt(x),
          val <- exp(x)+cos(x*y),
          val <- NA)
    return(val)
}

ys <- seq(1/3,3,leng=500)
xs <- seq(1/9,55/18-sqrt(109)/18,leng=500)

# creating the matrix of the values of the function
z <- sapply(ys, function(i) sapply(xs, function(j) func(j,i)))

contour(xs, ys, z,
        main="Integral and its domain",
        xlab="x", ylab="y",
        xlim=range(xs), ylim=range(ys),
        nlevels = 15,
        labcex = 0.75,
        lwd = 1)
    
    
#delimiting the domain
wd <- 2
lines(xs, domain_down(xs), lty=1, lwd=wd, col="skyblue")
lines(xs, domain_up(xs), col='skyblue', lty=1, lwd=wd)
lines(c(1/9,1/9), c(1/3,3), col='skyblue', lty=1, lwd=wd)
    
legend(1.5, .85,
       c("Function level","Domain limits"),  
       col=c("black", "skyblue"), 
       lty=1, lwd=1:wd, cex=0.8)
    
````

````r
# the function to integrate
func <- function(x,y) {
    return(exp(x)+cos(x*y))
}

# omega (the area over which the integral is defined) is delimited by the following points
p1 <- c(1/9,1/3)
p2 <- c(55/18-sqrt(109)/18, sqrt(55/18-sqrt(109)/18))
p3 <- c(2,3)
p4 <- c(1/9,3)

# so that we can draw points from the rectangle defined by (xlow,xhigh)*(ylow,yhigh)
xlow <- 1/9
xhigh <- 55/18-sqrt(109)/18
ylow <- 1/3
yhigh <- 3

# a function that allows drawing points from the given domain
draw_from_domain <- function(n) {
    #number of points is 2 times that required, just to make sure we actually get the number we're asking for
    num <- n*2
    #some boundaries
    xlow <- 1/9
    xhigh <- 55/18-sqrt(109)/18
    ylow <- 1/3
    yhigh <- 3
    #drawing the numbers from some uniforms
    xvals <- runif(num, xlow, xhigh)
    yvals <- runif(num, ylow, yhigh)

    #checking if the values are good
    mask <- c(1/3 <= 3*xvals & 3*xvals<=9-yvals & sqrt(xvals) <=yvals & yvals<=3)
    check <- (sum(mask)<n)
    ifelse(check,
        draw_from_domain(n),
        rightvals <- c(xvals[mask], yvals[mask]))
    
    return(rightvals[1:n])
}


# the actual calculation by means of a hit or miss procedure
calc <- function(n) {

    #drawing some points
    points <- draw_from_domain(n)
    xvals <- points[1]
    yvals <- points[2]
    fvals <- func(xvals, yvals)

    # the function is always positive in the given domain
    # actually, by looking at the contour plot, above, we can be sure that it does not go below 1
    # taking 1 as the lowest point grants some additional efficiency
    m <- 1
    M <- func(xhigh, sqrt(xhigh))
    
    #drawing random points in the interval spanned by the function
    comparison <- runif(n, min=m, max=M)

    const <- (M-m)*(xhigh-xlow)*(yhigh-ylow)
    val <- sum(ifelse(comparison<=fvals,
                     1,
                     0))
    val <- val*const/n
    return(val)
}

cat("The vaue of the integral is", calc(1e5))
````

This value makes sense, if one approximates the domain as a rectangle, with vertices $V_1(1/9,1.5), V_2(2.3,1.5), V_3(2.3,3), V_4(1/9,3)$.
